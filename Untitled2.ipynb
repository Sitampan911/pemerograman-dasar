```python
import pandas as pd
df = pd.read_csv('/content/sales.csv')

# Ringkasan statistik
df.describe()

# Rata-rata dari kolom 'quantity' (sebagai pengganti 'Population')
rata_rata_quantity = df['quantity'].mean()
print("Rata-rata quantity:", rata_rata_quantity)

# Modus dari kolom 'price_per_unit' (sebagai pengganti 'Airports')
modus_price_per_unit = df['price_per_unit'].mode()
print("Modus price_per_unit:", modus_price_per_unit)
```

```python
import matplotlib.pyplot as plt

df['quantity'].hist()
plt.title("Distribusi Kuantitas Penjualan")
plt.xlabel("Kuantitas")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Korelasi
# Menggunakan kolom yang ada di DataFrame df
corr_matrix = df[['quantity', 'price_per_unit']].corr()
print("Korelasi:\n", corr_matrix)

# Kovarian
# Menggunakan kolom yang ada di DataFrame df
cov_matrix = df[['quantity', 'price_per_unit']].cov()
print("Kovarian:\n", cov_matrix)
```

```python
# Menambah kolom kategori (menggunakan 'quantity' sebagai pengganti 'Airports')
df['Kategori_Kuantitas'] = df['quantity'].apply(lambda x: 'Tinggi' if x >= 2 else 'Rendah')
df

# Rata-rata 'price_per_unit' per kategori (menggunakan 'price_per_unit' sebagai pengganti 'Population')
group_mean = df.groupby('Kategori_Kuantitas')['price_per_unit'].mean()
print(group_mean)

# Agregasi lengkap (menggunakan 'price_per_unit' dan 'quantity' sebagai pengganti 'Population' dan 'Airports')
group_agg = df.groupby('Kategori_Kuantitas').agg({
    'price_per_unit': ['mean', 'max'],
    'quantity': ['mean', 'count']
})
group_agg
```
